```python
import pandas as pd
import numpy as np
```

```python
meals = pd.read_excel('meals.xlsx')


new_rows = pd.DataFrame({
  'MealName': ['Salmon Salad'],
  'MealTime': ['Lunch/Dinner'],
  'Calories': [np.NaN],
    'CookingTime': [0.5],
    'Protein': ['Salmon'],
    'Veggie': ['Lettuce/Avocado/Tomatos'],	
    'Carb'	: ['Potatoes'],
    'Dairy'	: ['Parm Cheese'],
    'Sauce' : ['White Sauce']
})



meals = pd.concat([meals, new_rows], ignore_index=True)



meals.to_excel('meals.xlsx', index=False)
meals.info()
meals.tail()
```
